```python
wine_reviews = pd.read_csv("winemag-data-130k-v2.csv", index_col=0)
wine_reviews.head(6)
```

```python
Q = wine_reviews.loc[(wine_reviews.price <= 400)]
Q.groupby([ 'price'])['points','title'].agg(max).reset_index()
```

```python
# có $400, tìm tên loại vang ngon nhất mà tôi có thể mua với số tiền đang có? 
O = wine_reviews.groupby(['price'])["points","title"].max().reset_index()
O[O.price<=400]
```

```python
# TEST 
wine_reviews.groupby(['variety','country'])['points', 'price' ].agg([max, min])
```

```python
# điểm vang cao nhất của dòng vang đó và giá tiền của chai thấp nhất cùng dòng.
wine_reviews.price = wine_reviews.price.fillna(0)
def highest_point_lowest_price(group):
    # print(group.price.isnull().sum()) # sanity check
    return pd.Series(data=[max(group.points), min(group.price),
                           group.title[group.price.idxmin()]], 
                     index=["highest points", "lowest price", "title"])

wine_reviews.groupby(['variety', 'country'])['points', 'price', 'title'].apply(highest_point_lowest_price)
```
